```python


# Estilo padr√£o para os gr√°ficos
sns.set(style="whitegrid", palette="pastel")
plt.rcParams["figure.figsize"] = (10, 5)

# Fun√ß√£o para gr√°fico de barras
def plot_bar(col, title="", xlabel="", ylabel="Frequ√™ncia", order=None):
    plt.figure()
    sns.countplot(x=col, data=df, order=order)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xticks(rotation=45)
    plt.show()

def plot_box(col, title="", prefixo_monetario=False):
    plt.figure()
    ax = sns.boxplot(x=df[col], color="skyblue")
    plt.title(title)

    # Quartis
    q1 = df[col].quantile(0.25)
    q2 = df[col].quantile(0.50)
    q3 = df[col].quantile(0.75)

    # Escolhe o formato com ou sem prefixo
    def formatar(valor):
        return f"R${valor:,.0f}" if prefixo_monetario else f"{valor:,.0f}"

    # Linhas verticais nos quartis
    for q, label in zip([q1, q2, q3], ['Q1 (25%)', 'Mediana (50%)', 'Q3 (75%)']):
        plt.axvline(q, color='red', linestyle='--')
        ax.annotate(f'{label}\n{formatar(q)}',
                    xy=(q, 0.3),
                    xycoords=('data', 'axes fraction'),
                    textcoords='offset points',
                    xytext=(0, 15),
                    ha='center',
                    fontsize=9,
                    color='red')

    plt.show()



# Fun√ß√£o para histograma com KDE
def plot_hist_kde(col, title=""):
    plt.figure()
    sns.histplot(df[col], kde=True)
    plt.title(title)
    plt.show()

# Fun√ß√£o para scatterplot entre duas vari√°veis
def plot_scatter(x, y, hue=None, title=""):
    plt.figure()
    sns.scatterplot(data=df, x=x, y=y, hue=hue)
    plt.title(title)
    plt.show()


def prever_vendas(tipo_midia: str, ano: int, mes: int, gasto: float) -> float:
    import pandas as pd
    import pickle

    # üîÅ Carrega modelo e encoder
    with open("modelo_regressao.pkl", "rb") as f:
        modelo = pickle.load(f)

    with open("encoder_midia.pkl", "rb") as f:
        encoder = pickle.load(f)

    # üßæ Monta DataFrame de entrada
    df_input = pd.DataFrame([{
        "Ano": ano,
        "Mes": mes,
        "Gastos com Publicidade (em R$)": gasto,
        "Tipo de Midia": tipo_midia
    }])

    # üß† Aplica mesma transforma√ß√£o do treino
    X_simulado = encoder.transform(df_input)

    # üìà Faz a previs√£o
    pred = modelo.predict(X_simulado)[0]

    print(f"üîé M√≠dia: {tipo_midia}")
    print(f"üìä Previs√£o: {pred:,.0f} mil unidades vendidas")
    return pred
```

### üõ†Ô∏è Pr√©-processamento para Regress√£o Linear

```python
# Vari√°veis
# Define vari√°vel alvo (target)
y = df["Previsao Inicial de Aumento de Vendas (em mil unidades)"]

# Define vari√°veis explicativas
X = df.drop(columns=["Previsao Inicial de Aumento de Vendas (em mil unidades)"])

# Define pr√©-processador
encoder = ColumnTransformer(
    transformers=[
        ("midia", OneHotEncoder(drop="first", handle_unknown="ignore", sparse_output=False), ["Tipo de Midia"])
    ],
    remainder="passthrough"
)
#drop="first",
# Aplica transforma√ß√£o
X_encoded = encoder.fit_transform(X)

# Recupera nomes das colunas ap√≥s transforma√ß√£o
nomes_dummies = encoder.named_transformers_["midia"].get_feature_names_out(["Tipo de Midia"])
nomes_finais = list(nomes_dummies) + [col for col in X.columns if col != "Tipo de Midia"]


# Cria DataFrame transformado para visualiza√ß√£o
df_transformado = pd.DataFrame(X_encoded, columns=nomes_finais)
print("Visualiza√ß√£o do DataFrame ap√≥s a transforma√ß√£o:\n")
df_transformado.head()
```

#### üîÑ Transforma√ß√£o de vari√°veis categ√≥ricas:
- A vari√°vel `Tipo de M√≠dia` foi convertida em vari√°veis dummy.

#### üéØ Defini√ß√£o das vari√°veis:
- `y`: Previs√£o Inicial de Aumento de Vendas (em mil unidades) ‚Äî vari√°vel-alvo.
- `X`: Todas as demais vari√°veis num√©ricas e dummies geradas ‚Äî vari√°veis explicativas.

> A matriz `X` agora est√° pronta para ser utilizada na modelagem.

```python
X = df_transformado

# Dividir em dados de treino e teste (80% treino, 20% teste)
X_train, X_test, y_train, y_test = train_test_split(df_transformado, y, test_size=0.2, random_state=42)

# Criar e treinar o modelo de regress√£o linear
model1 = LinearRegression()
model1.fit(X_train, y_train)

# Fazer previs√µes
y_pred = model1.predict(X_test)

# Calcular R¬≤ para medir o desempenho do modelo
r2 = r2_score(y_test, y_pred)
print(f"R¬≤ do modelo: {r2:.4f}")

# Salva modelo e encoder
with open("modelo_regressao.pkl", "wb") as f:
    pickle.dump(model1, f)

with open("encoder_midia.pkl", "wb") as f:
    pickle.dump(encoder, f)
```

```python
coeficientes = pd.DataFrame({"Vari√°vel": X.columns, "Coeficiente": model1.coef_})
coeficientes = coeficientes.sort_values(by="Coeficiente", ascending=False)

# Mostrar os coeficientes
coeficientes
```
